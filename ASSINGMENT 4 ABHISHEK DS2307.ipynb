```python
driver=webdriver.Chrome()
```

```python
driver.get('https://www.imdb.com/list/ls095964455/')
```

```python
serial=[]
try:
    sl=driver.find_elements(By.XPATH,'//div[@class="lister-item mode-detail"]')
    for i in sl:
        serial.append(i.text)
except:
    serial.append('NO ELEMENT')
```

```python
serial
```

```python
SERIAL_NAME = []
RATING= []
TIME = []

VOTES = []

for entry in serial:
    values=entry.split('\n')
    
    SERIAL_NAME.append(values[0])
    RATING.append(values[2])
    TIME.append(values[1])
    
    VOTES.append(values[6])
    

print(SERIAL_NAME)
print(RATING)
print(TIME)

print(VOTES)
```

```python
vote_list_cleaned = [vote.replace("Votes:", "") for vote in VOTES]

for vote in vote_list_cleaned:
    print(vote)
```

```python
daata=TIME
split_data = [element.split('|') for element in daata]


duration=[item[1] for item in split_data]
genre=[item[2] for item in split_data]


print(duration)
print(genre)
```

```python
series_names = []
years = []
for item in SERIAL_NAME:
    
    parts = item.split('(')
    
    name = parts[0].split('. ')[1]  
    year = parts[1].rstrip(')')  

    
    series_names.append(name)
    years.append(year)


print(series_names)
print(years)
```

```python
print(len(series_names))
print(len(years))
print(len(RATING))
print(len(duration))
print(len(genre))
print(len(vote_list_cleaned))
```

```python
df6=pd.DataFrame({'SERIAL':series_names,'YEAR':years,'RATING':RATING,'TIME':duration,'GENRE':genre,'VOTES':vote_list_cleaned})
df6
```

```python
clk5=driver.find_element(By.XPATH,'//select[@class="select-primary select select-sm rounded-full"]//option[5]')
clk5.click()
```

```python
zlink = []

start = 0
end = 27 

for page in range(start,end):
    Links=driver.find_elements(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"]')
    for i in Links:
        zlink.append(i.get_attribute('href'))
    
    next_button = driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[3]/div/button[2]')
    next_button.click()
    time.sleep(1)
```

```python
zlink
```

```python
len(zlink)
```

```python
detail = []

for i in zlink:
    driver.get(i)
    time.sleep(1)
   
    try:
        tab=driver.find_element(By.XPATH,'//p[@class="text-md"]')
        detail.append(tab.text)
    except NoSuchElementException:
        detail.append('NO_ELEMENT')
```

```python
len(detail)
```

```python
dataset_name = []

task = []
attribute_type = []
instances = []
no_attribute = []
date = []

for i in zlink:
    driver.get(i)
    time.sleep(1)
   
    try:
        typ=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[3]/p')
        task.append(typ.text)
    except NoSuchElementException:
        task.append('NO_ELEMENT')
    
    try:
        nam=driver.find_element(By.XPATH,'//h1[@class="text-3xl font-semibold text-primary-content"]')
        dataset_name.append(nam.text)
    except NoSuchElementException:
        dataset_name.append('NO_ELEMENT')
    
    try:
        f_type=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[4]/p')
        attribute_type.append(f_type.text)
    except NoSuchElementException:
        attribute_type.append('NO_ELEMENT')
        
    try:
        ins=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[5]/p')
        instances.append(ins.text)
    except NoSuchElementException:
        instances.append('NO_ELEMENT')
        
    try:
        feat=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[6]/p')
        no_attribute.append(feat.text)
    except NoSuchElementException:
        no_attribute.append('NO_ELEMENT')
        
    try:
        dat=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[1]/div[2]/h2')
        date.append(dat.text)
    except NoSuchElementException:
        date.append('NO_ELEMENT')
          
  
```

```python
date_list_cleaned = [date.replace("Donated on", "") for date in date]

for date in date_list_cleaned:
    print(date)
```

```python
df8=pd.DataFrame({'DATASET_NAME':dataset_name,'DATA_TYPE':detail,'TASK':task,'ATTRIBUTE_TYPE':attribute_type,'NO_OF_INSTANCES':instances,'NO_OF_ATTRIBUTES':no_attribute,'YEAR':date_list_cleaned})
df8
```
